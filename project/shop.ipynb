Funkcja dla wyświetlania listy brakujących wartości dla wszystkich kolumn w zbiorze danych i ich sumy

```python
def print_missing(df):
    missing = (df == "?").sum()
    print(missing)
    print(missing.sum())
```

Wczytanie pliku .xlsx z danymi,
oraz wyświetlenie listy brakujących wartości dla wszystkich kolumn w zbiorze danych i ich sumy

```python
import pandas as pd

df = pd.read_excel("../shopLog/logTest.xlsx")
print_missing(df)
df
```

```python
#informacje o kliencie wypełniane domyślnymi wartościami
df.loc[df["customerNo"] == "?", "customerNo"] = "unknown"
df.loc[df["maxVal"] == "?", "maxVal"] = 0
df.loc[df["customerScore"] == "?", "customerScore"] = 0
df.loc[df["accountLifetime"] == "?", "accountLifetime"] = 0
df.loc[df["payments"] == "?", "payments"] = 0
df.loc[df["address"] == "?", "address"] = "Mr"
df.loc[df["age"] == "?", "age"] = 0
df.loc[df["lastOrder"] == "?", "lastOrder"] = 0

print_missing(df)
```

```python
# jeżeli zamówienie doszło do skutku to musiało być dostępne
df.loc[df["order"] == "y", "availability"] = "completely orderable"
```

```python
print(df["onlineStatus"].unique())

#jako że w systemie nie istnieją wartości n przypuszcza się że system miał błąd zapisu danych przy statusie offline
df.loc[df["onlineStatus"] == "?", "onlineStatus"] = "n"
```

```python
# jeżeli zamówienie doszło do skutku, etap zamówienia musi być ostatni
df.loc[df["order"] == "y", "bStep"] = 5
```

```python
#jeżeli koszyk lub zbiór klikniętych produktów są puste to suma, minimalna i maksymalna cena tych produktów muszą być zerowe
df.loc[df["cCount"] == 0, "cMinPrice"] = 0
df.loc[df["cCount"] == 0, "cMaxPrice"] = 0
df.loc[df["cCount"] == 0, "cSumPrice"] = 0

df.loc[df["bCount"] == 0, "bMinPrice"] = 0
df.loc[df["bCount"] == 0, "bMaxPrice"] = 0
df.loc[df["bCount"] == 0, "bSumPrice"] = 0

print_missing(df)
```

```python
df.to_excel("../shopLog/result.xlsx")
```
